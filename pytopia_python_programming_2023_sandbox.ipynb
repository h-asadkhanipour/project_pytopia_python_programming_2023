# 15 Managing Python Packages with Conda

- when you create a new conda env without any packages determined, conda does not install  
a new version of python, by default it use system python in /usr/bin/python, after you install a package  
that requires a newer or older version of python, it will install that version of python.  
- you can run default version of python in a active conda env by typing exact version of system  
python, for example --> python3.12  

- **how to remove a env in conda**  
    - conda env remove -n env_name  
- conda install numpy=1.25 VS pip install numpy==1.25  

- **how to update a package installed in a conda env?**  
    - conda update PACKAGE_NAME  
    - to update all packages --> conda update --all  

- **env file --> YAML file extraction**  
    - conda env export > file_name.yml
- **for recreating a env from yml file**  
    - conda env create -f file_name.yml
