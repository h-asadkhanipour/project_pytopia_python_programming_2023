```python
f = open("test_file.md")
```

```python
content = f.read()
```

```python
f.close()
```

```python
print(content)
```

```python
with open("test_file.md") as f:
    for _ in range(3):
        content = f.readline()
```

```python
print(content)
```

```python
with open("test_file.md") as f:
    for _ in range(3):
        content += f.readline()
```

```python
print(content)
```

```python
# read files moree efficiently
with open("test_file.md") as f:
    for line in f:
        print(line, sep="")
```

```python
with open("test_file.md") as f:
    for line in f:
        print(line, end='')
```

# 03 Writing to Files

```python
f = open("test_file.md", mode='w')
```

```python
f.write("add new line")
```

```python
f.read()
```

```python
f.close()
```

```python
with open("test_file.md", mode='r') as f:
    for line in f:
        print(line, end='')
```

```python
f = open("test_file.md", mode="a+") 
```

```python
f.write("add second line\nadd third line \n")
```

```python
f.close()
```

```python
with open("test_file.md", mode='r') as f:
    for line in f:
        print(line, end='')
```

when you write sth to a file, file doesn't change instantly, changes are saved to RAM buffer,  
so you must move them from buffer to memory --> .flush() method do this task.
