```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        print(f"{self.name} is barking")
```

```python
class Account:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance
    
    def deposit(self, d_amount):
        self.__balance += d_amount
        print(f"{d_amount} has been transfered to {self.owner} account")
    
    def show_balance(self):
        print(f"{self.__balance} is in {self.owner} account")
    
    def withdraw(self, w_amount):
        if self.__balance >= w_amount:
            self.__balance -= w_amount
            print(f"{w_amount} has been withdrawed from {self.owner} account")
        else:
            print(f"There is not enough money in {self.owner} account")
```

```python
happy_account = Account("happy", 1000000)
```

```python
happy_account.owner
```

```python
happy_account.show_balance()
```

```python
happy_account.deposit(1000000)
```

```python
happy_account.withdraw(1234)
```

```python
happy_account.show_balance()
```

# 06 OOP Inheritance

Exercises

```python
class Employee:
    def __init__(self, name, email):
        self.name = name
        self.email = email
    
    def get_info(self):
        print(f"{self.name}\n{self.email}")
```

```python
class Managers(Employee):
    def __init__(self, name, email, department):
        super().__init__(name, email)
        self.department = department
    
    def get_info(self):
        super().get_info()
        print(f"{self.department}")
```

```python
employ_1 = Employee("name1", "email1")
```

```python
employ_1.get_info()
```

```python
manager_1 = Managers("name1", "email1", "department1")
```

```python
manager_1.get_info()
```
