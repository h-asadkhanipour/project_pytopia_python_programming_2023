# exercise

```python
def schedule_visit(section, time, activity="Feeding"):
    print(f"Planning to visit {section} section at {time} for {activity} activity.")
```

```python
schedule_visit("Reptiles", "10:00 AM")
```

```python
def add_special_request(section, request=None):
    if request == None:
        print(f"No special requests have been made for the {section} section.")
    else:
        print(f"Special request for the {section} section: {request}")
```

```python
add_special_request("Reptile")
```

```python
add_special_request("Mamals", "Wheelchair access")
```

# 06 Function Documentation

- docstring can be --> "", '', """ """

- exercises

```python
def calculate_rectangle_area(height: int, width: int) -> int:
    """This function calculates rectangle area

    :param height:  int
    :param width: int

    :return: int
    """
    area = height * width
    return area
```

```python
calculate_rectangle_area(12, 12)
```

```python
# 3
age = [14, 18, 21, 16, 30, 45, 17, 22, 61]
```

```python
list(filter(lambda age: age >= 18, age))
```

```python
# 4
prices = [19.99, 35.50, 89.99, 43.75]
list(map(lambda x: 0.9 * x, prices ))
```

```python
# 5
products = [('apple', 1.0), ('banana', 0.5), ('cherry', 1.5), ('date', 0.75)]
sorted(products, key=lambda item: item[1])
```

```python
# 6
books = [
    {'title': 'Book A', 'year': 2001},
    {'title': 'Book B', 'year': 1999},
    {'title': 'Book C', 'year': 2010},
    {'title': 'Book D', 'year': 1985}
]

sorted(books, key=lambda book: book['year'])
```
